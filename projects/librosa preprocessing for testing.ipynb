```python
audio_file_path = '../audio/test/mercy.wav'
```

```python
# playing the audio file

ipd.Audio(audio_file_path)
```

```python
# function for extracting the audio file

def feature_extractor(filename):
    audio, sample_rate = librosa.load(filename, res_type='kaiser_fast')
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T, axis=0)
    
    return mfccs_scaled_features
```

```python
voice = feature_extractor(audio_file_path)
```

```python
voice.shape
```

```python
# saving the file in a csv for modeling


np.savetxt('mercy.csv', voice, delimiter=',')
#np.savetxt('y_data.csv', y, delimiter=',')
```
